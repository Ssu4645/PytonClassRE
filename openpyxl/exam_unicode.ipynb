```python
# 파이썬 유니코드 인코딩 핵심 정리
# 문자열 컴퓨터 내부에서 숫자(유니코드)로 저장된다.
# 문자를 저장하는 숫자 체계를 유니코드라고 한다
text = '한'
print (text)
print (len(text))

# 문자를 유니코드로 변환
# string = 문자열
# char = 문자 ps : 정수도 가능
print (ord('가')) #문자 to 코드
print (chr(44032)) #코드 to 문자

#문자열을 byte로 변환
encoded = text.encode('utf-8')
print(encoded)
print(type(encoded))

#byte를 문자열로 변환
decoded = encoded.decode('utf-8')
print(decoded)
print(type(decoded))

#인코딩이 같지않으면 오류 발생
try:
    wrong = encoded.decode('cp949')
    print(wrong)
except UnicodeDecodeError:
    print('인코딩 방식이 올바르지 않습니다.')
```
