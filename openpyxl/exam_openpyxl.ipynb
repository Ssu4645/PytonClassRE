```python
from openpyxl import *

item1 = {
    '품목' : '바나나',
    '가격' : '3000',
    '수량' : '1개',
    '날짜' : '2026년 3월 12일'
}
item2 = {
    '품목' : '딸기',
    '가격' : '19200',
    '수량' : '30개',
    '날짜' : '2026년 3월 10일'
}
item3 = {
    '품목' : '사과',
    '가격' : '9000',
    '수량' : '3개',
    '날짜' : '2026년 3월 2일'
}
item4 = {
    '품목' : '신발',
    '가격' : '79000',
    '수량' : '1개',
    '날짜' : '2026년 3월 1일'
}
wb = Workbook()
ws = wb.active
ws['A1'] = '품목'
ws['B1'] = '가격'
ws['C1'] = '수량'
ws['D1'] = '날짜'
items = [item1, item2, item3, item4]

i = 2
for item in items:
    ws['A' + str(i)] = item['품목']
    ws['B' + str(i)] = item['가격']
    ws['C' + str(i)] = item['수량']
    ws['D' + str(i)] = item['날짜']
    i += 1
```

```python
from openpyxl import *
from openpyxl.styles import PatternFill
wb = Workbook()
ws = wb.active
ws['A1'] = '이름'
ws['B1'] = '국어'
ws['C1'] = '영어'
ws['D1'] = '수학'
color = PatternFill(fill_type = 'solid', fgColor ='0f84f2')
for cell in ws[1]:
    cell.fill = color
wb.save('bg_color.xlsx')
```

```python
from openpyxl import *
from openpyxl.styles import Font
wb = Workbook()
ws = wb.active
ws['A1'] = 'kkklo'
ws['A1'].font = Font(color = '0f84f2')
wb.save('font_color.xlsx')
print(ws['A1'].value)
```

```python
from openpyxl import *
from openpyxl.chart import BarChart, Reference

wb = Workbook()
ws = wb.active
ws.append(['Name', 'Score'])
ws.append(['홍길동', 90])
ws.append(['임꺽정', 20])
ws.append(['임꺽순', 70])
ws.append(['홍길순', 80])
# min_col = 2 B열
# min_row = 1 1행까지
# max_row = 5 5행까지
chart = BarChart() # 막대 차트 생성
data = Reference(ws, min_col = 2, min_row = 1, max_row = 5)
chart.add_data(data, titles_from_data = True) # titles_from_data  == 첫 행을 Series name으로 사용
chart.title = "그래프 제목"
ws.add_chart(chart, "E5")
wb.save('chart_exam.xlsx')
```
