```python
class Person():
    folder = []     #클래스 속성 : 모든 자식 클래스가 공유하는 요소
    def __init__(self, a):
        self.a = a    #인스턴스 속성 : 각각의 자식 클래스가 따로 가지는 요소
    def add_files(self, file):
        self.folder.append(file)

jin = Person(123)
jin.add_files('txt')

jinx = Person(456)
jinx.add_files('doc')

print ( jin.folder )
print ( jinx.folder )
print ( jinx.a )
print ( jin.a )
```
