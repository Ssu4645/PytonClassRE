```python
# 문제 : 함수를 사용해서 코드량을 확 줄여주세요.
def repeat_10times(x):
  dan = x
  i = 1
  print(f"== 구구단 {dan}단 출력 ==")
  while i <= 9:
    print(f"{dan} * {i} = {dan * i}")
    i += 1
userinp = "init"
while userinp.isdigit() == False:
    print("보고싶은 단을 입력해주세요")
    userinp = input("숫자를 입력해주세요") .strip()

    if userinp.isdigit() == True:
        repeat_10times(int(userinp))
    else:
        print("값은 숫자로만 입력해주세요")
```

```python
def isprime(num):
    count = 0
    for i in range(1, num + 1):
        if num % i == 0:
            count += 1
    if count == 2:
        print ("소수입니다")
    else:
        print ("소수가 아닙니다")
```
