```python

from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Edge()
driver.get('https://quotes.toscrape.com')

print(" ==== 1페이지 수집 시작 ====")

# 수집한 데이터를 저장할 리스트
results = []
quote_blocks = driver.find_elements(By.CLASS_NAME, 'quote')

for block in quote_blocks:
    # 각 불록 안에서 필요한 데이터 찾기
    quote_text = block.find_element(By.CLASS_NAME, 'text').text
    author = block.find_element(By.CLASS_NAME, 'author').text

    # 태그는 여러 개 일 수 있으므로 find_elements를 사용
    tag_elements = block.find_elements(By.CLASS_NAME, 'tag')

    tags = [t.text for t in tag_elements]
    # 리스트 컴프리헨션을 사용하여 각 tag 요소의 텍스트만 추출
    #  - 컴프리헨션 : 반복문을 이용해 새로운 컬렉션을 만드는 축약 문법
    #  - [표현식 for 변수 in 반복가능객체]
    # for t in tag_elements:
    #     tags.append(t.text)

    results.append({
        'page': 1,
        'quote': quote_text,
        'author': author,
        'tags': tags
    })

    print(f"{author} - {quote_text} / tag:{tags}")

print('현재까지 수집 개수 :', len(results))

# 브라우저 종료
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Edge()
driver.get('https://quotes.toscrape.com')

MAX_PAGES = 3
current_page = 1

while current_page <= MAX_PAGES:
    # CSS 문법을 이용해서 HTML요소를 찾는 방법
    # li.next a === class = next를 가진 li 안에 a 태그를 의미함
    next_buttons = driver.find_elements(By.CSS_SELECTOR, 'li.next a')
    if not next_buttons:
        print(" 더 이상 페이지가 없습니다")
        break
    next_buttons[0].click()
    current_page += 1
    print(f"====={current_page}페이지 수집 중 =====")
    quote_blocks = driver.find_elements(By.CLASS_NAME, 'quote')
    
    for block in quote_blocks:
        quote_text = block.find_element(By.CLASS_NAME, 'text').text
        author = block.find_element(By.CLASS_NAME, 'author').text
        tag_elements = block.find_elements(By.CLASS_NAME, 'tag')
        tags = [t.text for t in tag_elements]
        results.append({
            'page': current_page,
            'quote': quote_text,
            'author': author,
            'tags': tags
        })

    print(f"{author} - {quote_text} / tag:{tags}")
    print(f"====={current_page}페이지 수집 완료 =====")
    print('최종 수집 개수 : ', len(results))

    
```
