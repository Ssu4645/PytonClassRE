```python
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Edge()
driver.get("https://quotes.toscrape.com/")
div_elements = driver.find_elements(By.TAG_NAME, 'div')
print("TAG_NAME으로 찾은 div 태그 개수 : ", len(div_elements))
first_quote_text = driver.find_element(By.CLASS_NAME, 'text')
print('\n CLASS_NAME으로 찾은 첫 번째 명언 : ', first_quote_text.text)
# quote_container = driver.find_element(By.ID, 'quote')
# print('\n ID로 찾은 요소의 태그 이름 :', quote_container.tag_name)

#find_elements 여러개를 찾고 리스트 반환
#find_element 하나만 찾을때
one_quote_block = driver.find_element(By.CLASS_NAME, 'quote')
many_quote_block = driver.find_elements(By.CLASS_NAME, 'quote')
print('\n find_element로 찾은 quote불록 1개 :', type(one_quote_block))
print('\n find_elements로 찾은 quote블록 개수 :', len(many_quote_block))

# xPath
# 조건이나 경로로 정확하게 요소 선택 가능
# 페이지 구조가 바뀌면 깨질수 있다

xpath_quote_text = driver.find_element(
    By.XPATH,
    "//div[@class='quote']/span[@class='text']"
    #// = 문서 전체에서 검색
    #div[@class = 'quote'] = div 태그 중에서 class가 'quote 인요소
    #/ = 바로 아래 관계
    #span[@class = 'text'] = span 중에서 class가 text인요소
    #따라서 전체 문서에 있는 클래스가 quote인 div태그 아래에있는 클래스가 text인 span태그를 찾는다
)
print('\n XPATH로 찾은 첫번째 명언 : ', xpath_quote_text.text)
quote_first_block = driver.find_element(By.XPATH,"(//div[@class = 'quote'])[1]")
```
