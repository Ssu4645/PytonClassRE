```python
# 문제 1번임
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Edge()
driver.get("https://quotes.toscrape.com")
scaraped_results = []
target_block = driver.find_elements(By.CLASS_NAME, "quote")
for block in target_block:

    get_text = block.find_element(By.CLASS_NAME, "text").text
    get_author = block.find_element(By.CLASS_NAME, "author").text
    scaraped_results.append({
        get_author : get_text
    })
    print(f"{scaraped_results[-1]}\n")
driver.quit()
```
