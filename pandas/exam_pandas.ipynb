```python
# pandas 
# 데이터 분석과 처리를 위한 대표적인 라이브러리
import pandas as pd

# DataFrame = 여러개의 열로 구성된 표형태의 데이터
# 각 열에는 열 이름이 있음
# 각 행에는 자동으로 행 번호가 붙음
df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss Elizabeth",
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"],
    }
)
print(f"{df}\n")
#전체 출력
print(f"{df['Name']}\n")
print(df['Age'])
print(df['Sex'])

```

```python
# 이름이 age 인 시리즈 생성
ages = pd.Series([22, 35, 55], name = 'Age')
print(f"{ages}\n")
print(f"{df['Age']}\n")
# min
#print(f"최소 나이 = {}")
# max
```

```python
# DataFrame 에서 특정 행을 선택하는 방법 2가지
# - iloc    위치로 선택 (숫자 좌표로 접근)
# - loc     라벨로 선택 (인덱스에 붙어 있는 이름으로 접근)
import pandas as pd
df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss Elizabeth",
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"],
    }
)
print(f"{df}\n") # 전체 출력

print(f"{df.iloc[0]}\n") # 데이터 프레임의 첫번째 항목 불러오기

print(f"{df.loc[0]}\n") #인덱스가 지정되어 있지 않아서 일단은 0으로 불렀음
```

```python
import pandas as pd
df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss Elizabeth",
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"],
    }
)

# head(n)
# 상위 n개 출력
# tail(n)
# 하위 n개 출력
# 둘다 n이 없을 경우 기본적으로 5개 출력
print(f"상위 2개 요소 출력\n {df.head(2)}\n")
print(f"하위 2개 요소 출력\n {df.tail(2)}\n")
```

```python
# 조건 필터링
# boolean Indexing

#나이가 30이상인 사람만 확인하기
print(df[df['age']])
```

```python
#새로운 열 추가
df['Survived'] = [0, 1, 1]
#열 이름 바꾸기
df = df.rename(columns={'Sex' : 'gender'})
#열 삭제
df = df.drop('Survied', axis = 1)

```

```python
# 결측치 처리
# 값이 비어있는 상태
# pandas는 NaN 으로 표시한다
print(f"{df}")
print(f"{df.isnull}")
df2 = df.copy()
df2.loc[1, 'Age'] = None
print(f"{df2.isnull()}")

print(df2.dropna(axis = 0))     #결측치 지우기

print(f"{df.fillna(0)}")    #결측치 0으로 채우기


print(df2['Age'].fillna(df['Age'].mean))
#Age 열만 선택해서 fillna를 적용함
#NaN을 평균 값으로 대체

print(f"{df2.ffill()}\n") # 바로 이전 행의 값으로 NaN를 채운다
print(f"{df2.bfill()}\n") # 바로 다음 행의 값으로 NaN를 채운다
```
