```python
# 문제시작
a = 10
# `==` => 같다.
if ( a == 10 ):
  print("참1")
# `!=` => 같지 않다.
if ( a != 10 ):
  print("거짓1")
if ( a > 10 ):
  print("거짓2")
if ( a >= 10 ):
  print("참2")
b = 10
if ( a == b ):
  print("참3")
# c 에는 true or false 이 담긴다.
c = a != b
if ( c ):
  print("C 는 거짓3이야")
if ( c == False ):
  print("참4")
# `not` => 반전
if ( not c ):
  print("참5")
if ( not(not c) ):
  print("거짓4")
d = True
if ( c != d ):
  print("참6")
if ( 10 > 5 and -6 > -10 and 10 != 10 ):
  print("거짓5")
if ( 10 < 5 and -6 < -10 and 10 != 10 or 10 > 5 ):
  print("참7")
# 문제끝
```

```python
# 문제 : 10대, 20대, 30대 중 나이에 맞는 변수를 출력해주세요.
# range(시작, 끝, 증가값)
for i in range(1, 6):
    age = i*10+5
    print(f"Case{i} : age is {age}")
    if age >= 30 and age < 40:
        print ( "당신은 30대" )
    elif age < 30 and age >= 20: 
        print ("당신은 20대")
    elif age < 20 and age >= 10:
        print ("당신은 10대")
    else:
        print ("당신은 40대 이상 늙은이 입니다")
```

```python
# 문제 : 1부터 100 사이의 짝수만 출력해주세요.(개념 : while, if, %)
i =1
while i <= 100:
    a = i
    if a % 2 == 0 :
        print(i)
    i += 1
```
