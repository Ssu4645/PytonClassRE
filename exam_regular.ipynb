```python
import re
text = "안녕하세요. 저는 25살 같은 15살 같은 32살 입니다"
# r = row string \ (특수 문자)
# \d = 숫자
# + = 1개 이상
pattern = r"\d+"
result = re.findall(pattern, text)
print(result)
```

```python
#패턴 찾기
sample = "abc123xyz456"

print(re.findall(r'\d+', sample))
print(re.findall(r'[a-z]+', sample))
```

```python
#특정 형식 찾기
data = 'test123@gmail.com'
# \w : 문자 또는 숫자
# + : 한개 이상
# @ : 말 그대로 @
# w. : 점(특문이기 때문에 \를 붙임)
# 문자\숫자 + @ + 문자\숫자 + . + 문자\숫자
email_pattern = r'\w+@\w+\.\w+'
emails = re.findall(email_pattern, data)
print(emails)
```

```python
# 문자열 치환
# sub( 찾을 패턴, 바꿀 문자, 대상 문자열 )
test = "가격은 1200원 입니다."
new_test = re.sub(r'\d+', '', test)
print(new_test)
```

```python
# 정규식 vs 일반 문자열 처리
test = '몸무게: 50kg, 키: 170cm'
test.split(',')
nums = re.findall(r'\d+', test)
print(nums)
```
