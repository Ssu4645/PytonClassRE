```python
# 예제2
S = "안녕하세요."
print(" S 를 이루는 문자를 하나씩 출력, while")
i = 0
while i < len(S):
    print (S[i])
    i += 1
```

```python
# 예제3
S = "a안b녕c하d세e요f."
print("S 중에서 한글만 출력")
# for, if, or, continue, 방법1
for C in S:
    if C == 'a' or C == 'b' or C == 'c' or C == 'd' or C == 'e' or C == 'f':
        continue
    print(C)

```
